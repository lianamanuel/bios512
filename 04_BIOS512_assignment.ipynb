````r
mean_mf = df_marathon %>%
    filter(country %in% c('United States', 'Kenya', 'Ethiopia', 'Japan')) %>%
    group_by(MF) %>%
    summarize(race.time = mean(race.time))

# df_marathon %>% summarize(min = min(year), max = max(year))

options(repr.plot.width = 7, repr.plot.height = 5, repr.plot.res = 300)

df_marathon %>% ggplot(., aes(x = year, y = race.time)) + 
    geom_point(color = 'gray', aes(shape = MF)) +
    geom_point(data = df_filtered, aes(color = country, shape = MF)) +
    theme(plot.background = element_rect(fill = 'gray'),
          panel.background = element_rect(fill = 'white', color = 'black'),
          panel.grid.major = element_line(color = 'gray')) + 
    geom_hline(data = mean_mf, mapping = aes(yintercept = race.time, linetype = MF)) + 
    labs(x = "Year of race", y = "Race times") +
    ggtitle("Boston marathon race times by gender", subtitle = "Years 1897 - 2019") + 
    theme(plot.title.position = "plot")
````
