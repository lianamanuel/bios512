````r
options(repr.plot.width = 8, repr.plot.height = 3)

penguins_long %>% ggplot(., aes(x = species, y = value, color = sex)) + 
    geom_point(position = position_jitter(h = 0, w = 0.3), alpha = 0.5) + 
    facet_wrap(~name, scale = 'free_y', ncol = 4) + 
    stat_summary(fun = median, 
                 fun.min = median, 
                 fun.max = median, 
                 geom = "crossbar", 
                 width = 0.75, 
                 size = 0.2,
                 mapping = aes(group = species)) + 
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
````
